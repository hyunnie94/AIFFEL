#### 얼굴 크롭하기

```python
import face_recognition
import os

def get_cropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    if len(face_locations) == 0 :
        return []
    for face_location in face_locations:
        a, b, c, d = face_locations[0]
        cropped_face = image[a:c, d:b, : ]
    
    return cropped_face
```

#### 각 사진들의 임베딩 벡터 구하기

```python
image_file = os.path.join(dir_path, '이현경1.png')
face = get_cropped_face(image_file)   # 얼굴 영역을 구하는 함수
# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
embedding
```

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_cropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0 :
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
          
    return embedding_dict
```

tkinter : Tcl/Tk에 대한 파이썬 Wrapper로써 Tcl/Tk를 파이썬에 사용할 수 있도록 한 Lightweight GUI 모듈이다.  
Tcl : Tool Command Language의 약자.일종의 프로그래밍 언어.   
Tk : 크로스 플랫폼에 사용되는 일종의 GUI 툴킷   

```python
from tkinter import *      # 파이썬에 기본 내장된 Tkinter 패키지
import tkinter.font
import PIL
from PIL import ImageTk as itk
from PIL import Image


def get_image(name):
    image_file = os.path.join(dir_path, name+'.png')
    face = get_cropped_face(image_file)
    embeddidng = face_recognition.face_encodings(face)
    #embedding_dict = get_face_embedding_dict(dir_path)
    
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))

    img = Image.open(dir_path+'/'+sorted_faces[1][0]+'.png')
    img = img.resize((img.size[0]//3, img.size[1]//3), Image.ANTIALIAS)
    result_image = itk.PhotoImage(img)
    result = Label(lower_frame, image = result_image)
    #result_label.place(relwidth=1, relheight=1)
    result.image = result_image
    result.place(relwidth=1, relheight=1)
    
    return result
```
